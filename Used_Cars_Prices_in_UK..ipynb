```python
import pandas as pd

df = pd.read_csv("/content/sample_data/Project week 8/used_cars_UK.csv")
df.head()
```

```python
df1=df['Previous Owners'].value_counts()
df1.idxmax()
```

```python
import numpy as np
df['Previous Owners'].replace(np.nan,2,inplace=True)
```

```python
df1=df['Doors'].value_counts()
df1.idxmax()
df['Doors'].replace(np.nan,5,inplace=True)

df1=df['Seats'].value_counts()
df1.idxmax()
df['Seats'].replace(np.nan,5,inplace=True)

df1=df['Emission Class'].value_counts()
df1.idxmax()
df['Emission Class'].replace(np.nan,'Euro 5',inplace=True)

df1=df['Engine'].value_counts()
df1.idxmax()
df['Engine'].replace(np.nan,'1.6L',inplace=True)
```

```python
df.drop('Service history',axis=1, inplace=True)
```

```python
df.drop('Unnamed: 0',axis=1, inplace=True)
```

```python
df
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
x_variable = 'Body type'
y_variable = 'Price'

plt.figure(figsize=(8, 6))
plt.bar(df[x_variable], df[y_variable], alpha=0.5)
plt.title('BarPlot: ' + x_variable + ' vs ' + y_variable)
plt.xlabel(x_variable)
plt.ylabel(y_variable)
plt.grid(True)
plt.legend(labels=y_variable, loc='upper right')

plt.show()
```

```python
from matplotlib import pyplot as plt
df.plot(kind='scatter', x='Registration_Year', y='Previous Owners', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
TARGET = "Price"
tmp_df = df.groupby("Registration_Year").agg({TARGET:"mean"})
sns.lineplot(x=tmp_df.index, y=tmp_df[TARGET])
```

**Chart 5:- Line Plot [It shows Registration years by Price]**

- From this chart we see that price increases from year 2010 to 2023 as car models might be are new.
- And we see that in the year 1950's the price was high because for price (target variable) has many independent variable affecting it. It may be because it may be has only 1 previous owner like factors may affect the price

```python
sns.boxplot(x=df["Previous Owners"],y=df[TARGET])
```
